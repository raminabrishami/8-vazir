```python
import random
```

```python
class genetic():
    def __init__(self):
        self.genaration()
        self.crossover()
        self.mutation()
        self.evalution()
        
    def genaration(self):
        global kol
        global genarat
        global numb
        numb=0
        kol=[]
        genarat=[]
        binery=[1,1,1,1,1,1,1,1]
        for j in range(20):
            for i in range(8):
                binery[i]=random.choice(range(1,9))
            d=binery.copy()
            genarat.append(d)
        kol.append(genarat.copy())
        numb=1
       
        
    def crossover(self):
        global cros_over
        f=range(1,7)
        cross=[]  
        cros_over=[]
        
        for i in range(0,16,2):
            cut=random.choice(f)
            cross1=genarat[i][:cut]
            cross2=genarat[i][cut:]
            cross.append([cross1,cross2])
            cross1=genarat[i+1][:cut]
            cross2=genarat[i+1][cut:]
            cross.append([cross1,cross2])
    
        for j in range(0,16,2):
            cros=(cross[j][0]+cross[j+1][1])
            cros1=(cross[j+1][0]+cross[j][1])
            cros_over.append(cros)
            cros_over.append(cros1)
        kol.append(cros_over.copy()) 

        

    def mutation(self):
        global mutationn
        mutationn=[]
        f=range(8)
        f1=range(1,9)
        for i in range(6):
            rand=random.choice(f)
            rand1=random.choice(f1)
            a=True
            while a==True:
                if cros_over[i][rand]!=rand1:
                    cros_over[i][rand]=rand1
                    mutationn.append(cros_over[i])
                    a=False
                else:
                    rand1=random.choice(f1)
        kol.append(mutationn.copy())  
      
            
    
    def evalution(self):
        global genarat
        global emtiyaz
        global max_emtiyaz
        global kol1
        kol1=[]
        max_emtiyaz=[]
        genarat=[]
        emtiyaz=[]
        if numb==0:
            kol1=kol[0]+kol[1]+kol[2]
        else:
            kol1=kol[0]+kol[1]

        for l in range(len(kol1)):
            emtiyaz.append(30)
        for i in range(len(kol1)):
            for j in range(8):
                for k in range(8):
                    c=kol1[i][j]-kol1[i][k]
                    c=abs(c)
                    h=j-k
                    h=abs(h)
                    if j==k :
                        continue

                    elif c==0:
                        emtiyaz[i]-=1

                    elif c==h:
                        emtiyaz[i]-=1

        genarat=[]
        for i in range(20):
            c=emtiyaz[0]
            m=0
            for j in range(len(emtiyaz)):
                if c<emtiyaz[j]:
                    c=emtiyaz[j]
                    m=j
            l=kol1.pop(m)
            genarat.append(l)
            e=emtiyaz.pop(m)
            max_emtiyaz.append({e:m})
```
