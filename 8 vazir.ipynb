```python
import random
```

```python
class genetic():
    def __init__(self):
        self.genaration()
        self.crossover()
        self.mutation()
        self.evalution()
        
    def genaration(self):
        global genarat
        genarat=[]
        binery=[1,1,1,1,1,1,1,1]
        for j in range(20):
            for i in range(8):
                binery[i]=random.choice(range(1,9))
            d=binery.copy()
            genarat.append(d)

        
    def crossover(self):
        global cros_over
        f=range(1,7)
        cross=[]
        cros_over=[]
        cut=random.choice(f)
        
        for i in range(len(genarat)):
            cross1=genarat[i][:cut]
            cross2=genarat[i][cut:]
            cross.append([cross1,cross2])
        q=len(genarat)
        for j in range(0,q,2):
            cros=(cross[j][0]+cross[j+1][1])
            cros1=(cross[j+1][0]+cross[j][1])
            cros_over.append(cros)
            cros_over.append(cros1)
            


    def mutation(self):
        f=range(8)
        f1=range(1,9)
        rand=random.choice(f)
        for i in range(len(cros_over)):
            rand=random.choice(f)
            rand1=random.choice(f1)
            a=True
            while a==True:
                if cros_over[i][rand]!=rand1:
                    cros_over[i][rand]=rand1
                    a=False
                else:
                    rand1=random.choice(f1)
            
            
    def evalution(self):
        global genarat
        global emtiyaz
        global max_emtiyaz
        max_emtiyaz=[]
        genarat=[]
        emtiyaz=[]
        for l in range(len(cros_over)):
            emtiyaz.append(30)
        for i in range(len(cros_over)):
            for j in range(8):
                for k in range(8):
                    c=cros_over[i][j]-cros_over[i][k]
                    c=abs(c)
                    h=j-k
                    h=abs(h)
                    if j==k :
                        continue

                    elif c==0:
                        emtiyaz[i]-=1

                    elif c==h:
                        emtiyaz[i]-=1

        for i in range(10):
            c=emtiyaz[0]
            m=0
            for j in range(len(emtiyaz)):
                if c<emtiyaz[j]:
                    c=emtiyaz[j]
                    m=j
            l=cros_over.pop(m)
            genarat.append(l)
            e=emtiyaz.pop(m)
            max_emtiyaz.append({e:m})
    
```
